```python
df_import = pd.read_csv('USImport.csv', delimiter=';')
```

```python
# get 30 samples grouped by Country
n_countries = len(df_import['Country of Origin'].unique())
print(n_countries)
n_sample = 30 // n_countries
print(n_sample)
```

```python
df_sample_import = df_import.groupby('Country of Origin').apply(lambda x: x.sample(n=n_sample) if len(x) >= n_sample else x)
df_sample_import.head()
```

```python
steps = len(df_import) // 30
df_sample_import_step = df_import[::steps]
df_sample_import_step.head()
```

```python
print('Number of Exporters:', len(df_import['Exporter'].unique()))
print('Number of Importers:', len(df_import['Importer'].unique()))
print('Number of Countries of Origin:', len(df_import['Country of Origin'].unique()))
print('Number of Products:', len(df_import['Products'].unique()))
```

```python
print(len(df_import))
```

```python
importers = df_import['Importer'].unique().tolist()
sampled_importers = []
df_sample_import_custom = pd.DataFrame()
n_rows = 0
while n_rows < 30:
    # get random Importer
    chosen_importer = np.random.choice(importers)
    importers.remove(chosen_importer)
    sampled_importers.append(chosen_importer)
    n_rows += len(df_import[df_import['Importer'] == chosen_importer])

df_sample_import_custom = df_import[df_import['Importer'].isin(sampled_importers)]
df_sample_import_custom.head()
print(len(df_sample_import_custom))
```

```python
df_sample_import_custom = df_sample_import_custom[:30]
print(len(df_sample_import_custom))
```
