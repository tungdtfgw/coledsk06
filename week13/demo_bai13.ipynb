```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

```python
n_employees = 50
salary_std = 3
salary_mean = 18
salary_samples = np.random.normal(size=n_employees, scale=salary_std, loc=salary_mean)
```

```python
print(np.mean(salary_samples))
```

```python
plt.hist(salary_samples, bins=10, edgecolor='black')
plt.title('Histogram of Employee Salaries')
plt.xlabel('Salary (in thousands)')
plt.ylabel('Number of Employees')
plt.show()
```

```python
confidence = 0.95
alpha = 1 - confidence # 0.25
z_score = stats.norm.ppf(1 - alpha/2)
print(z_score)
margin = z_score * salary_std / n_employees ** 0.5
ci_lower = salary_mean - margin
ci_upper = salary_mean + margin
print(f"95% Confidence Interval for the mean salary: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
n_workers = 20                      # n
times = np.random.normal(size=n_workers, scale=2, loc=10)
time_mean = np.mean(times)          # Z
time_std = np.std(times, ddof=1)    # s

confidence = 0.95
alpha = 1 - confidence               # 0.05
t_score = stats.t.ppf(1 - alpha/2, df=n_workers - 1)
margin = t_score * time_std / n_workers ** 0.5
ci_lower = time_mean - margin
ci_upper = time_mean + margin
print(f"Mean times: {time_mean:.2f}")
print(f"Standard Deviation of times: {time_std:.2f}")
print(f"95% Confidence Interval for the mean time: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
n_customers = 200
n_customers_happy = 160
p_hat = n_customers_happy / n_customers
print(f"Sample Proportion of Happy Customers: {p_hat:.2f}")
confidence = 0.99
alpha = 1 - confidence               # 0.01
z_score = stats.norm.ppf(1 - alpha/2)
print(z_score)
margin = z_score * ((p_hat * (1 - p_hat)) / n_customers) ** 0.5
ci_lower = p_hat - margin
ci_upper = p_hat + margin
print(f"99% Confidence Interval for the proportion of happy customers: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
mean_declare = 7.5
# H0: Avarage math score = 7.5
# H1: Avarage math score != 7.5
```

```python
n_students = 25
math_scores = np.random.normal(size=n_students, scale=1.0, loc=7.8)
mean_math = np.mean(math_scores)
std_math = np.std(math_scores, ddof=1)
print(f"Mean Math Score: {mean_math:.2f}")
print(f"Standard Deviation of Math Scores: {std_math:.2f}")
```

```python
alpha = 0.01
t = (mean_math - mean_declare) / (std_math / n_students ** 0.5)
p_value = 2 * (1 - stats.t.cdf(abs(t), df=n_students - 1))

print(f"T-statistic: {t:.2f}")
print(f"P-value: {p_value:.4f}")
```

```python
if p_value <= alpha:
    print('Có bằng chứng cho thấy điểm trung bình toán khác 7.5 (bác bỏ H0)')
else:
    print('Không có bằng chứng cho thấy điểm trung bình toán khác 7.5 (không bác bỏ H0)')
```

```python
n_workers = 20
n_products = 75 # declare >= 75 products

products = np.array([78, 74, 80, 73, 77, 76, 79, 81, 72, 75, 74, 78, 77, 76, 80, 79, 73, 75, 78, 77])
mean_products = np.mean(products)
std_products = np.std(products, ddof=1)
print(f"Mean Products Produced: {mean_products:.2f}")
print(f"Standard Deviation of Products Produced: {std_products:.2f}")
```

```python
alpha = 0.05
# H0: mean_products <= 75
# H1: mean_products > 75
t = (mean_products - n_products) / (std_products / n_workers ** 0.5)
p_value = 1 - stats.t.cdf(t, df=n_workers - 1)
print(f"T-statistic: {t:.2f}")
print(f"P-value: {p_value:.4f}")
if p_value <= alpha:
    print('Có bằng chứng cho thấy sản phẩm trung bình > 75 (bác bỏ H0)')
else:
    print('Không có bằng chứng cho thấy sản phẩm trung bình > 75 (không bác bỏ H0)')
```

```python
weight = 500 # declare
std_weight = 3

n_samples = 100

weights = np.random.normal(size=n_samples, scale=std_weight, loc=500)
mean_weight = np.mean(weights)
std_weight_sample = np.std(weights, ddof=1)
print(f"Mean Weight: {mean_weight:.2f}")
print(f"Standard Deviation of Weights: {std_weight_sample:.2f}")
```

```python
# H0: mean_weight < 500
# H1: mean_weight >= 500
z = (mean_weight - weight) / (std_weight_sample / n_samples ** 0.5)
p_value = 1 - stats.norm.cdf(z)

alpha = 0.05
z_statistic = stats.norm.ppf(1 - alpha)
print(f"Z-statistic: {z:.2f}")
print(f"P-value: {p_value:.4f}")
```
