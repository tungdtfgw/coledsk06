```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import scipy.stats as stats
```

```python
hours = np.array([1, 2, 2, 3, 5, 4, 6, 8, 7, 9])
scores = hours * 10 + np.random.normal(0, 5, size=hours.shape)
print("Hours studied:", hours)
print("Scores obtained:", scores)
```

```python
df = pd.DataFrame({'Hours': hours, 'Scores': scores})
df.head()
```

```python
r = stats.pearsonr(df['Hours'], df['Scores'])[0]
print(f"Pearson correlation coefficient: {r:.2f}")
p_value = stats.pearsonr(df['Hours'], df['Scores'])[1]
print(f"P-value: {p_value:.4f}")
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Hours', y='Scores', data=df, s=100)
plt.title('Scatter Plot of Hours Studied vs. Scores Obtained')
plt.xlabel('Hours Studied')
plt.ylabel('Scores Obtained')
plt.show()
```

```python
price_chances = np.array([-20, -15, -10, -5, 0, 5, 10, 15, 20])
sales_volumes = 1000 - 2 * price_chances + np.random.normal(0, 30, size=price_chances.shape)
df = pd.DataFrame({'Price_Changes': price_chances, 'Sales_Volumes': sales_volumes})
df.head()
```

```python
r = stats.pearsonr(df['Price_Changes'], df['Sales_Volumes'])[0]
print(f"Pearson correlation coefficient: {r:.2f}")
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Price_Changes', y='Sales_Volumes', data=df, s=100)
plt.title('Scatter Plot of Price Changes vs. Sales Volume')
plt.xlabel('Price Changes')
plt.ylabel('Sales Volume')
plt.show()
```
