```python
name = input('Enter your name: ')    # input function always returns a string
age = int(input('Enter your age: '))  # convert the input string to an integer
print('Name:', name)
print('Age:', age)
print(type(name), type(age))
```

```python
age = age + 5
print(age)
```

```python
x = float(input('Enter a float number: '))
print('You entered:', x)
```

```python
print('Hello')
print('Hello', 'World')
```

```python
print(name, age, x)
```

```python
print(name, age, x, sep='+')
```

```python
# format output
print(f'Name: {name}, Age: {age}, x: {x}')
```

```python
s = 'Hello'
a = 5
pi = 3.1415926
print(f'|{s:12}|')      # reserve 12 spaces for string
print(f'|{a:12}|')      # reserve 12 spaces for integer
print(f'|{pi:12}|')     # reserve 12 spaces for float
```

```python
print(f'|{s:>12}|')      # right aligned
print(f'|{a:^12}|')      # center aligned
print(f'|{pi:<12}|')     # left aligned
```

```python
# only for float numbers
print(f'|{pi:12.2f}|')     # 2 decimal places
print(f'|{pi:12.4f}|')     # 4 decimal places
```
