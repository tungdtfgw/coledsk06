```python
# Operations for integer: + - * // % **
a = 10
b = 2
print(f'{a} + {b} = {a + b}')
print(f'{a} - {b} = {a - b}')
print(f'{a} * {b} = {a * b}')
print(f'{a} // {b} = {a // b}') # integer division
print(f'{a} / {b} = {a / b}')   # float division
print(f'{a} % {b} = {a % b}')   # modulus
print(f'{a} ** {b} = {a ** b}') # exponentiation
```
