```python
import pandas as pd
import numpy as np
from apyori import apriori
```

```python
df_retail = pd.read_csv('Online_Retail_5000.csv')
df_retail.head()
```

```python
df_retail.info()
```

```python
df_retail['StockCode'].nunique()
```

```python
df_retail['InvoiceNo'] = df_retail['InvoiceNo'].astype(str) + '_'
df_retail['StockCode'] = df_retail['StockCode'].astype(str) + '_'
```

```python
transactions = []
invoices = df_retail['InvoiceNo'].unique()
print(len(invoices))
```

```python
for iv in invoices:
    items = df_retail[df_retail['InvoiceNo'] == iv]['StockCode'].tolist()
    transactions.append(items)

print(len(transactions))
```

```python
print(transactions[0:10])
```

```python
apr_rules = apriori(transactions, min_support=0.02, min_confidence=0.3, min_lift=3, min_length=2)
apr_rules = list(apr_rules)
```

```python
apr_rules_sorted = sorted(apr_rules, key=lambda x: x.support, reverse=True)
for i, rule in enumerate(apr_rules_sorted[:5]):
    print(f"Rule {i+1}:")
    print(f"Items: {rule.items}")
    print(f"Support: {rule.support}")
    for ordered_stat in rule.ordered_statistics:
        print(f'  {ordered_stat.items_base} -> {ordered_stat.items_add}')
        print(f"  Confidence: {ordered_stat.confidence}")
        print(f"  Lift: {ordered_stat.lift}")
    print("=====================================")
```
