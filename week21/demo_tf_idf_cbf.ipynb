```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movie_data = pd.read_csv('movies_metadata.csv')  
```

```python
movie_data.head()
```

```python
movie = movie_data[['overview', 'title']].dropna()
```

```python
movie.head()
```

```python
movie.info()
```

```python
movie_overviews = movie['overview']
movie_titles = movie['title']
```

```python
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(movie_overviews)
```

```python
print(X.shape)
```

```python
col_names = vectorizer.get_feature_names_out()
print(X[0, :].toarray())
```

```python
movie_index = pd.Series(movie.index, index=movie['title']).drop_duplicates()
print(movie_index)
```

```python
watched_movie = 'Century of Birthing'
watched_movie_index = movie_index[watched_movie]
print(watched_movie_index)
watched_movie_overview = movie_overviews[watched_movie_index]
print(watched_movie_overview)
watched_movie_vector = vectorizer.transform([watched_movie_overview])
```

```python
similarities = cosine_similarity(watched_movie_vector, X).tolist()[0]
print(len(similarities))
print(similarities)
```

```python
for i, score in enumerate(similarities):
    print(f"index: {i}, score: {score}")
```

```python
sorted_similar_movies = sorted(list(enumerate(similarities)), key=lambda x: x[1], reverse=True)
print(sorted_similar_movies[:6])
```

```python
recommend_indexes = [i[0] for i in sorted_similar_movies[0:6]]
for index in recommend_indexes:
    print(movie_titles.iloc[index])
```
