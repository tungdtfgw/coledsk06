```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras.models import Sequential
```

```python
import matplotlib.pyplot as plt
```

```python
X = np.linspace(0, 10, 200)
Y = 3.0 * X + np.random.normal(0, 2, 200)

plt.scatter(X, Y)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
# định nghĩa model
model = Sequential(
    [
        Dense(1, input_shape=(1,), activation=None)
    ]
)
```

```python
# cấu hình model
model.compile(optimizer="sgd", loss="mse")
```

```python
# huấn luyện model
model.fit(X, Y, epochs=10, verbose=1)
```

```python
loss_values = model.history.history['loss']
plt.plot(loss_values)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.show()
```

```python
Y_pred = model.predict(X)
plt.scatter(X[::10], Y[::10], color='blue')
plt.scatter(X[::10], Y_pred[::10], color='red')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
W, b = model.layers[0].get_weights()
print(W, b)
print(f'Y = {W[0][0]:.2f} * X + {b[0]:.2f}')
```

```python
X = np.linspace(-20, 20, 200)
Y = 2.0 * X ** 2 + 5.0 * X - 10 + np.random.normal(-25, 25, 200)

plt.scatter(X, Y)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
# định nghĩa model
model = Sequential(
    [
        Dense(1, input_shape=(1,), activation="tanh")
    ]
)
# cấu hình model
model.compile(optimizer="sgd", loss="mse")

# huấn luyện model
model.fit(X, Y, epochs=100, verbose=1)

# vẽ loss
loss_values = model.history.history['loss']
plt.plot(loss_values)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.show()
```

```python
Y_pred = model.predict(X)
plt.scatter(X[::10], Y[::10], color='blue')
plt.scatter(X[::10], Y_pred[::10], color='red')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
model = Sequential(
    [
        Dense(32, input_shape=(1,), activation="relu"),
        Dense(32, activation="relu"),
        Dense(32, activation="relu"),
        Dense(1, activation=None),
    ]
)
# cấu hình model
model.compile(optimizer="adam", loss="mse")

# huấn luyện model
model.fit(X, Y, epochs=200, verbose=1)

# vẽ loss
loss_values = model.history.history['loss']
plt.plot(loss_values)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.show()
```

```python
Y_pred = model.predict(X)
plt.scatter(X[::10], Y[::10], color='blue')
plt.scatter(X[::10], Y_pred[::10], color='red')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
