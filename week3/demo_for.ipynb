```python
s = 0
n = int(input('Enter n: '))
for i in range(1, n + 1):
    s = s + i # s += i
print(f'Sum 1-{n} = {s}')
```

```python
n = int(input('Enter n: '))
s = 0
for _ in range(n):
    a = int(input('Enter a number: '))
    s += a # s = s + a
print(f'Sum of {n} numbers = {s}')
```

```python
n = int(input('Enter n: '))
count = 0
for _ in range(n):
    a = int(input('Enter a number: '))
    if a % 2 == 0:
        # increse count
        count += 1
print(f'Count of even numbers = {count}')
```

```python
n = int(input('Enter n: '))
for i in range(1, 11):
    print(f'{i:2} x {n} = {i * n:2}')
```
