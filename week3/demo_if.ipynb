```python
# Nhập 1 ngày trong tuần và in ra là ngày làm việc hay cuối tuần
day = int(input('Enter a day in the week: '))
if day < 1 or day > 7:
    print('Invalid day')
elif 1 < day < 7:
    print('Working day')
else:
    print('Weekend')
```
