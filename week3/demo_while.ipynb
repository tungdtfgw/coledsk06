```python
answer = 'yes'
while answer == 'yes':
    n = int(input('Enter n: '))
    # print multiplication table of n
    for i in range(1, 11):
        print(f'{i:2} x {n} = {i * n:2}')
    
    answer = input('Do you want to continue? (yes/no): ')

print('Goodbye!')
```
