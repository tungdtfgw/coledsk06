```python
# function with return value
def add(a, b):
    c = a + b
    return c
```

```python
x = 5
y = 10
z = add(x, y)
print(f'{x} + {y} = {z}')
```

```python
def swap(a, b):
    c = a
    a = b
    b = c
    print(f'a = {a}, b = {b} inside function swap')

a = 5
b = 10
swap(a, b)
print(f'a = {a}, b = {b} outside function swap')
```

```python
def print_line(length):
    for _ in range(length):
        print('*', end=' ')
    print()  # print new line
```

```python
print_line(5)
```

```python
print_line(10)
```

```python
n = int(input('Enter n: '))
for i in range(n):
    print_line(i + 1)
```

```python
# new version of print_line
def print_line(length, c):
    for _ in range(length):
        print(c, end=' ')
    print()  # print new line
```

```python
print_line(5, '?')
print_line(5, '#')
```

```python
def print_triangle(nlines, c):
    for i in range(nlines):
        print_line(i + 1, c)
```

```python
print_triangle(5, '@')
```

```python
print_triangle(4, '$')
```

```python
def print_line(length, c, end_line='\n'):
    for _ in range(length):
        print(c, end=' ')
    print(end_line, end='')  # print new line or custom end
```

```python
n = 5
for i in range(n):
    print_line(n - i - 1, ' ', end_line='')
    print_line(2 * i + 1, '*', end_line='\n')
```
