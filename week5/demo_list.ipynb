```python
def indexes(a, x):
    found = []
    for i in range(len(a)):
        if a[i] == x:
            found.append(i)
    return found
```

```python
found = indexes(a, 2)
print(found)  # Output: [0, 2, 6]
```

```python
# extend method
a = [1, 2, 3]
b = [4, 5, 6]
a.extend(b)
print(a)  # Output: [1, 2, 3, 4, 5, 6]
```
