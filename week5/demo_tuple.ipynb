```python
a = [1, 2, 3, 4, 5] # list
b = (1, 2, 3, 4, 5) # tuple
```

```python
print(a[0])
print(b[0])
```

```python
a[0] = 10
print(a)
```

```python
b[0] = 10  # This will raise a TypeError
print(b)
```

```python
a.append(6)
print(a)  # Output: [10, 2, 3, 4,
```

```python
b.append(6)  # This will raise an AttributeError
print(b)
```

```python
b = (1, 2, 3) # tuple
b = b * 2
print(b)
```

```python
s = "hello"
print(s * 2)
```

```python
s[0] = 'H'  # This will raise a TypeError
```

```python
s = s * 2
print(s)
```

```python
a = (1, 2)
b = (3, 4)
c = a + b
print(c)  # Output: (1, 2, 3, 4)
```

```python
# packing and unpacking
a = 1, 2, 3  # packing
print(a)  # Output: (1, 2, 3)
```

```python
m, n, p = a  # unpacking
print(m)  # Output: 1
print(n)  # Output: 2
print(p)  # Output: 3
```

```python
a = 4
b = 10
# swap #1
c = a
a = b
b = c
print(a, b)  # Output: 10 4
```

```python
a = 4
b = 10
# swap #1
a = b + a
b = a - b
a = a - b
print(a, b)  # Output: 10 4
```

```python
a = 4
b = 10
a, b = b, a # packing and unpacking
print(a, b)  # Output: 10 4
```
