```python
nhap_lieu = True
list_products = []

while nhap_lieu:
    product = input('Nhập tên sản phẩm: ')
    price = int(input('Nhập giá sản phẩm: '))

    list_products.append((product, price))

    tiep_tuc = input('Bạn có muốn tiếp tục không? (y/n): ')
    nhap_lieu = tiep_tuc == 'y'
```

```python
for product, price in list_products:
    if price > 20000:
        print(f'Sản phẩm: {product}, Giá: {price} VND')
```

```python
def merge_lists(a, b):
    c = []
    i = 0 # pointer for list a
    j = 0 # pointer for list b
    while i < len(a) and j < len(b):
        if a[i] < b[j]:
            c.append(a[i])
            i += 1
        else:
            c.append(b[j])
            j += 1
    # append remaining elements
    if i < len(a):  # still have elements in a
        for k in range(i, len(a)):
            c.append(a[k])
    elif j < len(b):  # still have elements in b
        for k in range(j, len(b)):
            c.append(b[k])
        
    return c
```

```python
a = [1, 5, 10, 20]
b = [2, 9, 11, 15, 25]
c = merge_lists(a, b)
print(c)
```

```python
votes = [1, 2, 3, 4, 4, 1, 1, 2, 2, 1]
vote_count = {}
for v in votes:
    if v not in vote_count:
        vote_count[v] = 1   # first vote
    else: # from second vote onwards
        vote_count[v] += 1  # increment count
for vote, count in vote_count.items():
    print(f'Ứng cử viên {vote} nhận được {count} phiếu bầu')
```
