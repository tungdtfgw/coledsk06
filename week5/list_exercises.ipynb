```python
import random as rd

numbers = rd.sample(range(100), 10)
print(numbers)
```

```python
s = 0
for n in numbers:
    s += n

print(s)
```

```python
s = 0
n = len(numbers)
for i in range(n):
    s += numbers[i]
print(s)
```

```python
fruits = ['banana', 'mango', 'kiwi', 'apple']
prices = [100, 200, 50, 150]

max_price = max(prices)
i = prices.index(max_price)
print(fruits[i])
```

```python
print(fruits[prices.index(max(prices))])
```

```python
max_price = prices[0]
i_max = 0
for i in range(len(prices)):
    if prices[i] > max_price:
        max_price = prices[i]
        i_max = i
print(fruits[i_max])
```

```python
# find fruit that price > 100
for i in range(len(prices)):
    if prices[i] > 100:
        print(fruits[i])
```

```python
count = 0
for fruit in fruits:
    if 'a' in fruit:
        print(fruit)
        count += 1

print(f'There are {count} fruits contains letter a')
```
