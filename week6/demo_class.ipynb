```python
class Student:
    # Constructor (hàm khởi tạo), 2 tham số: name và age
    def __init__(self, name, age):
        # gan giá trị tham số cho thuộc tính của đối tượng
        self.name = name
        self.age = age
    
    def hi(self):
        print(f'My name is {self.name}, I am {self.age} years old.')
```

```python
john = Student('John', 20) # create an object of Student class
john.hi()               # call method hi of object john
```

```python
print(john.name)
print(john.age)
```

```python
paul = Student('Paul', 22)
print(paul.name)
print(paul.age)
paul.hi()
```

```python
paul.age = -20
paul.hi()
```

```python
class Student:
    def __init__(self, name, age):
        self.name = name    # name is a public attribute
        self.__age = age    # __age is a private attribute
    
    def hi(self):
        print(f'My name is {self.name}, I am {self.__age} years old.')
```

```python
john = Student("John Doe", 21)
print(john.name)  # name is accessible because it's public
```

```python
print(john.__age)
```

```python
john.hi()
```

```python
class Student:
    def __init__(self, name, age):
        self.name = name    # name is a public attribute
        self.__age = age    # __age is a private attribute
    
    def get_age(self):
        return self.__age
    
    def set_age(self, new_age):
        if new_age < 18:
            print("Age cannot be less than 18.")
        else:
            self.__age = new_age
    
    def hi(self):
        print(f'My name is {self.name}, I am {self.__age} years old.')
```

```python
john = Student("John Doe", 21)
print(john.name)  # Output: John Doe
print(john.get_age())  # Output: 21
```

```python
john.set_age(22)
john.hi()  
```

```python
john.set_age(-22)
john.hi()
```

```python
class Employee:
    def __init__(self, eid, name):
        self.__id = eid
        self.__name = name

    def get_id(self):
        return self.__id
    
    def get_name(self):
        return self.__name
```

```python
john = Employee(1, 'John Lennon')
paul = Employee(2, 'Paul McCartney')
print(john.get_id(), john.get_name())
print(paul.get_id(), paul.get_name())
```

```python
class Employee:
    def __init__(self, name, salary):
        self.set_name(name)
        self.set_salary(salary)
    
    def get_name(self):
        return self.__name

    def get_salary(self):
        return self.__salary

    def set_name(self, name):
        if name == "":
            print("Name cannot be empty.")
        else:
            self.__name = name
    
    def set_salary(self, salary):
        if salary < 0:
            print("Salary cannot be negative.")
        else:
            self.__salary = salary
```

```python
john = Employee('John', -2000)
print(john.get_name())
print(john.get_salary())
```

```python
class Employee:
    def __init__(self, name, salary):
        self.__name = name
        self.__salary = salary

    @property
    def name(self):
        return self.__name
    
    @name.setter
    def name(self, new_name):
        if new_name == "":
            print("Name cannot be empty.")
        else:
            self.__name = new_name

    @property
    def salary(self):
        return self.__salary

    @salary.setter
    def salary(self, new_salary):
        if new_salary < 0:
            print("Salary cannot be negative.")
        else:
            self.__salary = new_salary
```

```python
john = Employee('John', 50000)
print(john.name)
john.salary = 30000
print(john.salary)
```
