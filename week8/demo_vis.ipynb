```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
x = np.linspace(0, 10, 100)
```

```python
y = np.sin(x)
```

```python
plt.plot(x, y)
plt.title("Sine Wave")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
y = 4 * x
plt.plot(x, y)
plt.title("y = 4x")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
products = ['Iphone 17', 'Samsung Galaxy S30', 'Google Pixel 10', 'OnePlus 12', 'Sony Xperia 1 V']
sales = [150, 120, 90, 80, 70]
plt.bar(products, sales, color=['blue', 'green', 'red', 'purple', 'orange'])
plt.title("Smartphone Sales in 2025")
plt.xticks(rotation=45)
plt.xlabel("Products")
plt.ylabel("Units Sold (in thousands)")
plt.show()
```

```python
plt.pie(sales, labels=products, autopct='%1.0f%%', explode=[0.1, 0.1, 0, 0, 0])
plt.title("Smartphone Sales Distribution in 2025")
plt.show()
```

```python
reading_scores = np.random.normal(75, 10, 200)
writing_scores = reading_scores + np.random.normal(0, 5, 200)
plt.scatter(reading_scores, writing_scores, alpha=0.6)
plt.title("Reading vs Writing Scores")
plt.xlabel("Reading Scores")
plt.ylabel("Writing Scores")
plt.show()
```

```python
x = np.linspace(1, 10, 10, dtype=int)
y = 2*x + 5

plt.plot(x, y, marker='o', linestyle='-.', color='g', ms=5)
plt.title("Line Plot of y = 2x + 5")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
x = np.random.normal(0, 10, 100)
plt.hist(x, bins=15, color='skyblue', edgecolor='black')
plt.title("Histogram of Normally Distributed Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
# redraw using seaborn for better aesthetics
sns.histplot(x, bins=15, kde=True, color='skyblue')
plt.title("Histogram of Normally Distributed Data with KDE")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
df_penguin = pd.read_csv('penguins_simple.csv')
df_penguin.head()
```

```python
df_penguin.describe()
```

```python
df_penguin.isnull().sum()
```

```python
df_penguin.dropna(inplace=True)
```

```python
df_penguin.isnull().sum()
```

```python
bill_length = df_penguin['bill_length_mm']
flipper_length = df_penguin['flipper_length_mm']

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.scatter(bill_length, flipper_length, color='red', alpha=0.5)
plt.title("Bill Length vs Flipper Length")
plt.xlabel("Bill Length (mm)")
plt.ylabel("Flipper Length (mm)")

plt.subplot(1, 2, 2)
bill_depth = df_penguin['bill_depth_mm']
plt.scatter(bill_length, bill_depth, color='blue', alpha=0.5)
plt.title("Bill Length vs Bill Depth")
plt.xlabel("Bill Length (mm)")
plt.ylabel("Bill Depth (mm)")

plt.show()
```

```python
df_penguin_numbers = df_penguin[['bill_length_mm', 
                                'bill_depth_mm', 
                                'flipper_length_mm', 
                                'body_mass_g']]

penguin_corr = df_penguin_numbers.corr()
sns.heatmap(penguin_corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix of Penguin Measurements")
plt.show()
```

```python
body_mass = df_penguin['body_mass_g']
plt.scatter(flipper_length, body_mass, color='purple', alpha=0.6)
plt.title("Flipper Length vs Body Mass")
plt.xlabel("Flipper Length (mm)")
plt.ylabel("Body Mass (g)")
plt.show()
```

```python
plt.scatter(bill_depth, flipper_length, color='darkblue', alpha=0.6)
plt.title("Bill Depth vs Flipper Length")
plt.xlabel("Bill Depth (mm)")
plt.ylabel("Flipper Length (mm)")
plt.show()
```

```python

# Scatter plot bill length vs bill depth, mi o mt mu
sns.scatterplot(data=df_penguin, x='bill_length_mm', y='bill_depth_mm', hue='island', s=50, alpha=0.6)
plt.title("Bill Length vs Bill Depth (by Island)")
plt.xlabel("Bill Length (mm)")
plt.ylabel("Bill Depth (mm)")
plt.show()
```

```python
mean_body_mass = df_penguin.groupby('island')['body_mass_g'].mean()
mean_body_mass.head()
```

```python
mean_body_mass.plot(kind='bar', color='blue')
plt.title("Average Body Mass of Penguins by Island")
plt.xlabel("Island")
plt.ylabel("Average Body Mass (g)")
plt.show()
```

```python
# Count penguins by species on Gentoo island
gentoo_species_count = df_penguin[df_penguin['island'] == 'Gentoo']['species'].value_counts()
print(gentoo_species_count)
```

```python
# Check what islands exist
print("Unique islands:")
print(df_penguin['island'].unique())
print("\nSpecies distribution by island:")
print(df_penguin.groupby('island')['species'].value_counts())
```

```python
# 3 pie charts for gender distribution by species
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Filter each species separately
adelie_sex = df_penguin[df_penguin['species'] == 'Adelie']['sex'].value_counts()
chinstrap_sex = df_penguin[df_penguin['species'] == 'Chinstrap']['sex'].value_counts()
gentoo_sex = df_penguin[df_penguin['species'] == 'Gentoo']['sex'].value_counts()

# Plot each pie chart
axes[0].pie(adelie_sex, labels=adelie_sex.index, autopct='%1.0f%%', startangle=90)
axes[0].set_title('Adelie - Gender Distribution')

axes[1].pie(chinstrap_sex, labels=chinstrap_sex.index, autopct='%1.0f%%', startangle=90)
axes[1].set_title('Chinstrap - Gender Distribution')

axes[2].pie(gentoo_sex, labels=gentoo_sex.index, autopct='%1.0f%%', startangle=90)
axes[2].set_title('Gentoo - Gender Distribution')

plt.tight_layout()
plt.show()
```
